```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import random
%matplotlib inline
```

```python
salary = pd.read_csv('./salary.csv', encoding='SHIFT-JIS')
salary.head()
```

```python
# グラフ
x = salary['年齢']
y = salary['年収']

plt.plot(x,y)
plt.grid(color ='0.8')
plt.show()
```

```python
# 差分グラフ
cnt = len(salary)

# 差分を取る
diff_y = []

for i in range(0, cnt-1):
    diff_y.append(y[i+1]-y[i])

x = salary['年齢']
y = salary['年収']

plt.plot(x[1:],diff_y)
plt.grid(color ='0.8')
plt.show()
```

```python
x = np.arange(-10,10,0.1)

# 元の関数
y = x ** 3 + 3*x **2 + 3*x + 1 
plt.plot(x,y)
plt.grid(color ='0.8')
plt.show()

# 導関数
y2 = 3*x**2 + 6*x + 3
plt.plot(x,y2)
plt.grid(color ='0.8')
plt.show()

# 導関数
y2 = 6*x + 6
plt.plot(x,y2)
plt.grid(color ='0.8')
plt.show()
```

```python
x = np.arange(-10000,10000,0.1)

# 元の関数
y = -x ** 3 + 100*x **2 - 2500*x + 5000 
plt.plot(x,y)
plt.grid(color ='0.8')
plt.show()

# 導関数
y2 = -3*x**2+200*x+2500
plt.plot(x,y2)
plt.grid(color ='0.8')
plt.show()

# 導関数の微分
y3 = -6*x+200
plt.plot(x,y3)
plt.grid(color ='0.8')
plt.show()
```

```python
# 積分
def calc_area(dx):
    h = 0
    area = 0
    cnt = int(10/dx)
    for i in range(1,cnt+1):
        h = i * dx
        s = h * dx
        area += s
    return area
```

```python
calc_area(1)
```

```python
calc_area(0.5)
```

```python
calc_area(0.01)
```
