```python
# 連立方程式
a = Symbol('a')
b = Symbol('b')

ex1= a + b -1
ex2 = 5*a + b -3


print(solve((ex1,ex2)))
```

```python
# 直行する交点

x = np.arange(-1, 6)
y = 1/2 * x + 1/2
y2 = -2 * x + 7

plt.plot(x,y,marker='o')
plt.plot(x,y2,marker='o')

plt.grid(color='0.8')
plt.show()
```

```python
# ２直線の交点
# 式の定義
x = Symbol('x')
y = Symbol('y')

ex1= -3/2 * x +6 -y
ex2 = 1/2 * x +2 -y

# 回答
print(solve((ex1,ex2)))
```

```python
# 垂直２等分線
# 基となる線分の傾きと切片
a1 = (5-1)/(6-0)
b1 = 1

# 線分の中点
cx = (0+6)/2
cy = (1+5)/2

# 線分に直行する直線の傾き(a2 = -1/a1)
a2 = -1 /a1

# 線分に直行する直線の切片(b2 = y -a2*x)
b2 = cy - a2 * cx

# 直線の式
x = np.arange(0, 7)
y = a1*x + b1
y2 = a2 * x + b2

plt.plot(x,y,marker='o')
plt.plot(x,y2,marker='o')
plt.axis('equal')
plt.grid(color='0.8')
plt.show()
```

```python
# 三角比を使って円を描く

# 角度
th = np.arange(0,360)

# 円の半径
r = 5

# 円周上の点Pの座標
x = r * np.cos(np.radians(th))
y = r * np.sin(np.radians(th))

plt.plot(x,y)
plt.axis('equal')
plt.grid(color='0.8')
plt.show()
```

```python
# 三角比を使って円を描く

# 円の半径
r = 300

# 円周上の点Pの座標
x = np.arange(-r , r+1)
y = np.sqrt(r**2 - x**2)

plt.plot(x,y)
plt.axis('equal')
plt.grid(color='0.8')
plt.show()
```

```python
# 三角比を使って円を描く

# 円の中心
a = 200
b = 300

# 円の半径
r = 300

# 円周上の点Pの座標
x = np.arange(a-r , a+r+1)
y = np.sqrt(r**2 - (x-a)**2)+b
y2 = -y + 2*b

plt.plot(x,y)
plt.plot(x,y2)

plt.axis('equal')
plt.grid(color='0.8')
plt.show()
```

```python
# 三角形の面積
# ヘロンの公式
x = [1,3,6]
y = [5,1,4]

a = math.sqrt((x[1]-x[0])**2 + (y[1]-y[0])**2)
b = math.sqrt((x[2]-x[1])**2 + (y[2]-y[1])**2)
c = math.sqrt((x[2]-x[0])**2 + (y[2]-y[0])**2)

s = (a+b+c) /2
math.sqrt(s*(s-a)*(s-b)*(s-c))
```

```python
import math
 
def heron(a, b, c):
    d = (a + b + c) / 2
    return math.sqrt(d * (d - a) * (d - b) * (d - c))
 
print("A:", heron(12, 15, 9))
```
